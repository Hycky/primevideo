```python
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Carregando os dados
df = pd.read_csv('data/data.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Convertendo 'releaseYear' para inteiro
df['releaseYear'] = pd.to_numeric(df['releaseYear'], errors='coerce').astype('Int64')
```

```python
df = df.dropna().reset_index(drop=True)
display(df)
```

```python
# 1. Distribuição de tipos de conteúdo
plt.figure(figsize=(10, 6))
df['type'].value_counts().plot(kind='bar')
plt.title('Distribuição de Tipos de Conteúdo')
plt.xlabel('Tipo')
plt.ylabel('Contagem')
plt.show()
```

```python
# 2. Top 10 países com mais produções
plt.figure(figsize=(12, 6))
df['availableCountries'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Países com Mais Produções')
plt.xlabel('País')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. TOP 10 Distribuição de classificações
plt.figure(figsize=(10, 6))
df['imdbAverageRating'].value_counts().head(10).plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribuição de Classificações')
plt.axis('equal')
plt.show()
```

```python
# 4. Evolução do número de lançamentos ao longo dos anos
plt.figure(figsize=(15, 6))
df['releaseYear'].value_counts().sort_index().plot(kind='line')
plt.title('Evolução do Número de Lançamentos ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Número de Lançamentos')
plt.show()
```

```python
from wordcloud import WordCloud

plt.figure(figsize=(12, 8))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(df['genres'].unique()))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Nuvem de Palavras dos Gêneros do conteúdo do Amazom Prime')
plt.show()
```
